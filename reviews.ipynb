```python
from datetime import datetime
def date(X):
    days = []
    months = []
    years = []
    for i in X:
        day, month, year = tuple(i.split(" "))
        day, month, year = int(day), datetime.strptime(month, "%B").month, year
        days.append(day)
        months.append(month)
        years.append(year)
    return days, months

data["day"], data["month"],= date(data["review_date"])
data = data.drop("review_date", axis=1)
data.head()
```

```python
cat_features = [col for col in data.columns if data[col].dtypes == "object"]
num_features = [col for col in data.columns if data[col].dtypes in ["int64", "float64"]]
```

```python
for i in cat_features:
    data[i] = data[i].apply(str.lower)

data.head()
```

```python
text = data.review_body
```

```python
def wordnet_pos(X):
    if X.startswith == "J":
        return wordnet.ADJ
    elif X.startswith == "V":
        return wordnet.VERB
    elif X.startswith == "N":
        return wordnet.NOUN
    elif X.startswith == "R":
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
class lemmatization:
    def __init__(self):
        self.wnl = WordNetLemmatizer()
    def __call__(self, doc):
        token = word_tokenize(doc)
        word_pos = nltk.pos_tag(token)
        return [self.wnl.lemmatize(word, pos=wordnet_pos(tag)) for word, tag in word_pos]
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

vectorizer = TfidfVectorizer(stop_words='english',tokenizer=lemmatization(), lowercase=True)
```

```python
from sklearn.model_selection import train_test_split

text_tf = vectorizer.fit_transform(text)

t_train, t_test, y_train, y_test = train_test_split(text_tf, data["review_rating"], test_size=0.1, random_state=12)
t_train, t_valid, y_train, y_valid = train_test_split(t_train, y_train, test_size=0.1, random_state=42)
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=12)
t_train, y_train = smote.fit_resample(t_train, y_train)
```

```python
sns.histplot(y_train)
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()

model.fit(t_train, y_train)
```

```python
from sklearn.model_selection import GridSearchCV

params = {
    "alpha":[0.1,0.5,1.0,2.0]
}

grid_search = GridSearchCV(model, params, scoring="roc_auc", cv=5)
grid_search.fit(t_valid, y_valid)
```

```python
best_params = grid_search.best_params_
best_params
```

```python
model.set_params(alpha=best_params["alpha"])#, penalty=best_params["penalty"], solver=best_params["solver"])
```

```python
from sklearn.metrics import accuracy_score, roc_auc_score
prediction= model.predict(t_test)

print("Accuracy: ", accuracy_score(y_test, prediction))
```

```python
def testing(text):
    tokens = nltk.word_tokenize(text)
    text_vectorized = vectorizer.transform([' '.join(tokens)])
    prediction = model.predict(text_vectorized)[0]

    if prediction == 1:
        return "Positive sentiment"
    elif prediction == 0:
        return "Neutral sentiment"
    else:
        return "Negative sentiment"
```
